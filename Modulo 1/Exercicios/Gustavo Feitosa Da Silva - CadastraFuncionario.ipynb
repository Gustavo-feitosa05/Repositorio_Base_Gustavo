### üè¢ Desafio: Sistema de Cadastro de Funcion√°rios com Listas e Loop for

üìå Contexto do Desafio

Voc√™ foi contratado(a) pela **MSCorp** para desenvolver um sistema de cadastro de funcion√°rios usando listas. O programa deve:

‚úÖ Cadastrar alguns funcion√°rios (quantos o usu√°rio quiser)

‚úÖ Armazenar nome, cargo e sal√°rio de cada um em listas separadas.

‚úÖ No final, exibir um relat√≥rio organizado com todos os dados.

### üìã Requisitos do Sistema


1Ô∏è‚É£ Estrutura dos Dados
Usaremos 3 listas separadas para armazenar os dados:



```python
nomes = []
cargos = []
salarios = []
```

2Ô∏è‚É£ Fluxo do Programa


O sistema pergunta: "Quantos funcion√°rios deseja cadastrar?"

Usando um loop for, o programa repete o cadastro N vezes

Para cada funcion√°rio, coletar:

üìõ Nome completo

üíº Cargo

üíµ Sal√°rio (aceitar decimais, ex: 3500.50)

Ao final, exibir todos os funcion√°rios cadastrados em formato de tabela

### üéØ Exemplo de Sa√≠da

### Meu C√≥digo:



```python
# Listas para armazenar os dados
nomes = []
cargos = []
salarios = []

quantidade = int(input("Quantos funcion√°rios deseja cadastrar? "))

print("-----| ‚ú® Relat√≥rio de Funcion√°rio ‚ú®|-----")
print("")
print(f"Total de funcion√°rios: {quantidade}")
print("")

for i in range(quantidade):

    nome = input("Qual nome do funcionario??")
    cargo = input("Qual o cargo do funcionario??")
    salario = input("Qual o sal√°rio do funcionario")

    nomes.append(nome)
    cargos.append(cargo)
    salarios.append(salario)



    print("-----------------------------------------------------")
    print("Nome                | Cargo                  | Sal√°rio")
    print("-----------------------------------------------------")
    print(f"{nome:<20}| {cargo:<20}   | R$ {salario}")
    print("-----------------------------------------------------")
    print("")
    print("-"* 60)
```

### Graficos para impressionar

```python
import matplotlib.pyplot as plt

#=========== VISUALIZA√á√ÉO COM MATLOTLIB ===========

# Configura√ß√£o do estilo
plt.style.use('ggplot')
plt.figure(figsize=(12, 6))

# Gr√°fico 1: Distribui√ß√£o de Sal√°rios
plt.subplot(1, 2, 1)
plt.bar(nomes, salarios, color='skyblue')
plt.title('Sal√°rios dos Funcion√°rios')
plt.xlabel('Funcion√°rios')
plt.ylabel('Sal√°rio (R$)')
plt.xticks(rotation=90)
plt.grid(axis='y')

# Gr√°fico 2: Distribui√ß√£o por Cargo (contagem)
plt.subplot(1, 2, 2)
cargos_unicos = list(set(cargos)) # Remove duplicatas
contagem = [cargos.count(cargo) for cargo in cargos_unicos]
plt.pie(contagem, labels=cargos_unicos, autopct='%1.1f%%', colors=['lightcoral', 'lightgreen', 'gold'])
plt.title('Distribui√ß√£o por Cargo')

plt.tight_layout()
plt.show()

# Gr√°fico 3: Histograma de faixas salariais (extra)

plt.figure(figsize=(8, 5))
faixas = [0, 3000, 5000, 7000, 9000, 11000]
plt.hist(salarios, bins=faixas, edgecolor='black', color='lightseagreen')
plt.title('Distribui√ß√£o por faixa Salarial')
plt.xlabel('Faixas Salariais (R$)')
plt.ylabel('Quantidade de Funcion√°rios')
plt.xticks(faixas)
plt.show()
```
